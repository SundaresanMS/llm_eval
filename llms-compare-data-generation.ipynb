Price-to-Earnings (P/E) Ratio: Indicates how much investors are paying for a dollar of earnings. High values suggest optimism about future growth.
Debt-to-Equity (D/E) Ratio: Measures a company's financial leverage and is calculated by dividing its total liabilities by stockholders' equity. Lower values are generally preferred, indicating less risk.
Return on Equity (ROE): Shows how effectively management is using a company’s assets to create profits. Higher values are typically seen as positive.
Current Ratio: Indicates a company's ability to pay short-term obligations or those due within one year. A ratio over 1 suggests that the company is in good financial health.
Earnings Per Share (EPS): The portion of a company's profit allocated to each outstanding share of common stock, serving as an indicator of a company's profitability. Increasing EPS over time is a good sign.
